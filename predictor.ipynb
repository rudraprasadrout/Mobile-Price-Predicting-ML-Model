```python
import pandas as pd
```

```python
data =  pd.read_csv("Flipdata.csv")
```

```python
df = df.drop(columns=["Processor","Warranty"])
```

```python
df
```

```python
df['Brand'] = df['Title'].apply(lambda x:" ".join(x.split()[0:1]))
```

```python
df.head()
```

```python
df.drop(columns = ["Title"], inplace = True)
```

```python
df.sample(5)
```

```python
def get_digits_loop(text):
    result = ''
    for char in str(text):
        if char.isdigit():
            result += char
    return result
```

```python
df["Battery"] = df["Battery"].apply(get_digits_loop)
df["Prize"] = df["Prize"].apply(get_digits_loop)
```

```python
def get_digits_split_loop(text):
    text = str(text)
    numbers = []
    current_num = ""
    
    for char in text:
        if char.isdigit():
            current_num += char  
        else:
            if current_num:       
                numbers.append(current_num)
                current_num = "" 
                
    if current_num:              
        numbers.append(current_num)
    
    return " + ".join(numbers)
```

```python
df["Camera"] = df["Camera"].apply(get_digits_split_loop).drop(columns = ["Camera"])
```

```python
df
```

```python
import re
```

```python
def get_inches_only(text):
    match = re.search(r'(\d+(?:\.\d+)?)\s*inch', str(text), re.IGNORECASE)
    if match:
        return match.group(1)
    return None
```

```python
df["Display"] = df["Display"].apply(get_inches_only)
```

```python
df
```

```python
def extract_ram_rom(text):
    ram = None
    rom = None
    
    parts = str(text).split('|')
    
    for part in parts:
        part = part.strip() 
        
        if 'RAM' in part:
            match = re.search(r'\d+', part)
            if match:
                ram = match.group()
                
        elif 'ROM' in part:
            match = re.search(r'\d+', part)
            if match:
                rom = match.group()
    return pd.Series([ram, rom])
```

```python
df[['RAM', 'ROM']] = df['Storage'].apply(extract_ram_rom)
```

```python
df
```

```python
df = df.drop(columns = ["Storage"])
```

```python
df
```

```python
def sum_all_numbers(text):
    if pd.isna(text):
        return 0
    matches = re.findall(r'(\d+(?:\.\d+)?)', str(text))
    total_sum = sum(float(num) for num in matches)
    return int(total_sum) if total_sum.is_integer() else total_sum
```

```python
df['Camera_Score'] = df['Camera'].apply(sum_all_numbers)
```
